## 1.1.9

Функция $f(x) = \frac{1}{3}x^3 - 2.5x^2 + 6x + a$ — это кубическая функция. Для того чтобы она имела ровно 2 действительных корня, одно из них должно быть кратным, то есть двойным. Это происходит, когда график функции касается оси $x$ в одной точке, а в другой точке пересекает ось $x$. Для кратного корня выполняются следующие условия:
1. $f(x) = 0$ — значение функции в этой точке равно нулю.
2. $f'(x) = 0$ — это критическая точка, где касательная к графику горизонтальна (либо максимум, либо минимум).
3. $f''(x) \neq 0$ — вторая производная не равна нулю, что подтверждает касание оси $x$.

### Шаг 1: Найдём производную функции $f(x)$

Для нахождения критических точек, где могут находиться корни, найдём первую производную:

$f'(x) = \frac{d}{dx} \left( \frac{1}{3}x^3 - 2.5x^2 + 6x + a \right)$

$f'(x) = x^2 - 5x + 6$

### Шаг 2: Найдём дискриминант производной

Критические точки находятся там, где производная равна нулю. Чтобы функция имела кратный корень, уравнение $f'(x) = 0$ должно иметь один корень. Это возможно, когда дискриминант квадратного уравнения $f'(x) = 0$ равен нулю.

Для уравнения $f'(x) = x^2 - 5x + 6$, дискриминант равен:

$D = (-5)^2 - 4 \cdot 1 \cdot 6 = 25 - 24 = 1$

Поскольку дискриминант положительный ($D > 0$), это уравнение имеет два различных корня, а значит, функция $f(x)$ имеет две критические точки.

### Шаг 3: Найдём критические точки

Решим уравнение $f'(x) = 0$:

$x^2 - 5x + 6 = 0$

Решаем квадратное уравнение:

$x = \frac{5 \pm \sqrt{1}}{2} = \frac{5 \pm 1}{2}$

$x_1 = 3, \quad x_2 = 2$

Итак, критические точки — это $x_1 = 3$ и $x_2 = 2$.

### Шаг 4: Исследуем поведение функции в этих точках

Поскольку $f(x)$ является кубической функцией, её график имеет форму "S-образной" кривой. Чтобы функция имела ровно два корня, один из этих корней должен быть кратным, а это возможно, если в одной из критических точек функция $f(x)$ принимает значение ноль.

Найдём значения функции в критических точках:

$f(2) = \frac{1}{3}(2)^3 - 2.5(2)^2 + 6(2) + a = \frac{8}{3} - 10 + 12 + a = \frac{8}{3} + a + 2$

$f(2) = \frac{14}{3} + a$

$f(3) = \frac{1}{3}(3)^3 - 2.5(3)^2 + 6(3) + a = \frac{27}{3} - 2.5(9) + 18 + a = 9 - 22.5 + 18 + a$

$f(3) = 4.5 + a$

### Шаг 5: Найдём $a$, чтобы функция имела два действительных корня

Чтобы функция имела два корня, в одной из критических точек значение функции должно быть равно нулю. Рассмотрим оба случая.

#### Случай 1: $f(2) = 0$

$\frac{14}{3} + a = 0$

$a = -\frac{14}{3}$

#### Случай 2: $f(3) = 0$

$4.5 + a = 0$

$a = -4.5$

### Шаг 6: Выбор наименьшего по модулю значения $a$

У нас есть два значения $a$:
- $a = -\frac{14}{3} \approx -4.67$
- $a = -4.5$

Наименьшее по модулю значение — это $a = -4.5$.

### Ответ:
Наименьшее по модулю значение параметра $a$ — это $\boxed{-4.5}$.


## Особенности кубических уравнений

Кубическая функция $f(x) = \frac{1}{3}x^3 - 2.5x^2 + 6x + a$ может иметь:

- Три различных действительных корня,
- Два действительных корня (один из которых кратный),
- Один действительный корень (и два комплексных).

Когда говорят о том, что кубическая функция имеет ровно два действительных корня, подразумевается, что один из них является двойным (или кратным). То есть это точка, где график касается оси $x$, но не пересекает её (такой корень называют кратным).

### Как это происходит?

Кратный корень возникает, когда график функции касается оси $x$, а это значит, что в этой точке выполняются следующие условия:

1. $f(x) = 0$ — значение функции в этой точке равно нулю.
2. $f'(x) = 0$ — это критическая точка, где касательная к графику горизонтальна (либо максимум, либо минимум).
3. $f''(x) \neq 0$ — вторая производная не равна нулю, чтобы гарантировать, что точка действительно является касательной, а не пунктом перегиба.

Таким образом, чтобы кубическая функция имела ровно два действительных корня, необходимо, чтобы один из корней был кратным (двойным). То есть функция должна иметь один кратный корень и один простой корень.
